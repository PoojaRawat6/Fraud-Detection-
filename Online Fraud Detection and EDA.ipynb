VISUALIZATION AND EXPLORATORY ANALYSIS

```python
# Importing essential libraries for visualization: matplotlib for general plotting and seaborn for statistical graphics.

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#These are columns, here we have defined a list named 'feature' containing names of columns from DataFrame (df) that we want to visualize. 
feature=['step','amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']

#Now we will start a for loop that iterates over each element (i) in the feature list. 
#In each iteration, i will represent one of the column names from the feature list.

for i in feature:
    plt.xlabel(i) #Sets the label for the x-axis of the plot to the current value of i. In the first iteration, i will be 'step', then 'amount', and so on
    df[i].plot(kind='hist', bins=5, figsize=(4,2), facecolor='pink',edgecolor='black') #generates a histogram plot of the column i
    plt.show()  #Displays the plot for the current column (i). 
                #After displaying each plot, the loop iterates to the next column in feature and repeats the plotting process
```

```python
num_cols = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']

plt.figure(figsize=(12, 8))
sns.boxplot(data=df[num_cols])
plt.title('Boxplots of Numerical Columns')
plt.xlabel('Features')
plt.ylabel('# Plot boxplots for numerical columns nValues')
plt.xticks(rotation=45)
plt.show()
```

```python
#Quantile-based Flooring and Capping for outliers
for i in num_cols:
    print(i)
    print(df[i].quantile(0.10))
    print(df[i].quantile(0.90))

    print('\n')
```

```python
#removal of the outlier

for i in num_cols:
    lower = df[i].quantile(0.10)
    upper = df[i].quantile(0.90)
    df[i] = np.where(df[i] <lower, lower,df[i])
    df[i] = np.where(df[i] >upper, upper,df[i])
    print('num_cols: ',i)
    print('Skewness value: ',df[i].skew())
    print('\n')
```

```python
#Now we'll take a look at it after removing outliers
for i in num_cols:
    plt.xlabel(i)
    df[i].plot(kind='hist', bins=5, figsize=(12,6), facecolor='pink',edgecolor='black')
    plt.show()
```

Transforming the categorical features into numerical

```python
# Map transaction types to numerical values for easier analysis and modeling
df["type"] = df["type"].map({"CASH_OUT": 1, "PAYMENT": 2, "CASH_IN": 3, "TRANSFER": 4, "DEBIT": 5})

# Map the isFraud column to more readable string labels
df["isFraud"] = df["isFraud"].map({0: "No Fraud", 1: "Fraud"})

# Print the first few rows of the dataframe to verify the changes
print(df.head())
```

```python
# splitting the data

x = np.array(df[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(df[["isFraud"]])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# training a machine learning model

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.10, random_state=42)
model = DecisionTreeClassifier()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

```python
features = np.array([[1, 8900.2, 8990.2, 0.0]])
print(model.predict(features))
```
